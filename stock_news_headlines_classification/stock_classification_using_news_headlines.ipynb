```python
# Libraries
import pandas as pd
import nltk
#nltk.download()

from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Read CSV
df = pd.read_csv('Data.csv',encoding = 'ISO-8859-1')
```

```python
# Splitting Data based on Date
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']

x_train = train.iloc[:,2:]
y_train = train['Label']
x_test = test.iloc[:,2:]
y_test = test['Label']
```

```python
# Converting Data to Lowercase

for i in x_train.columns:
    x_train[i] = x_train[i].str.lower()
```

```python
# Renaming Columns
x_train.columns = [i for i in range(len(x_train.columns))]
```

```python
# Removing Punctuations For Train Data

x_train.replace('[^a-zA-Z]',' ',regex=True,inplace=True)
```

```python
# Joining Data of All Columns

Headlines = []
for i in range(len(x_train)):
    Headlines.append(' '.join(str(i) for i in x_train.iloc[i,:]))
```

```python
# Tokenizing Text
token_text = []
for i in Headlines:
    token_text.append(nltk.word_tokenize(i))
```

```python
# Lemmatization

Lemmatizer = WordNetLemmatizer()

lemma_text = []
for i in token_text:
    word_lemma = []
    for j in i:
        word_lemma.append(Lemmatizer.lemmatize(j))
    lemma_text.append(word_lemma)
```

```python
# Joining Text after Lemmatization

x_train = []
for i in lemma_text:
    x_train.append(' '.join(i))
```

```python
# Joining Test Data Columns

Headlines = []
for i in range(len(x_test)):
    Headlines.append(' '.join(str(i) for i in x_test.iloc[i,:]))
    
x_test = Headlines
```

```python
# Bag of Words

countvector = CountVectorizer(ngram_range=(2,2))
train_set = countvector.fit_transform(x_train)
```

```python
# Random Forest Classifier Model Building

random_forest_clf = RandomForestClassifier(n_estimators=200, criterion='entropy')
random_forest_clf.fit(train_set,y_train)
```

```python
# Vectorizing Test Data

test_data = countvector.transform(x_test)
```

```python
# Predictions
predictions = random_forest_clf.predict(test_data)
```

```python
# Performance Metrices

matrix = confusion_matrix(y_test,predictions)
print(matrix)
print()
acc_score = accuracy_score(y_test,predictions)
print('Accuracy Score: ',acc_score)
print()
clf_report = classification_report(y_test,predictions)
print(clf_report)
```
